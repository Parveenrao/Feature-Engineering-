```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df['number'].unique()
```

```python
df['number'].value_counts().plot(kind = 'bar')
plt.title("Passenger Travelling With")
plt.show()
```

```python
# extract numerical_column
df['number_numerical'] = pd.to_numeric(df['number'] , errors = 'coerce' , downcast = 'integer')
```

```python
# extract categorical_columns
df['number_categorical'] = np.where(df['number_numerical'].isnull() , df['number'] , np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')   # extract numerical part
df['cabin_cat'] = df["Cabin"].str[0]       # extract categorical part
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind = 'bar')
```

```python
df['ticker_num'] = df['Ticket'].apply(lambda s : s.split()[-1])
```

```python
df
```

```python
df['ticker_num'] = pd.to_numeric(df['ticker_num'],
                                   errors='coerce',
                                   downcast='integer')
df
```

```python
df['ticket_cat'] = df['Ticket'].apply(lambda s : s.split()[0])
df
```

```python
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit() , np.nan , df['ticket_cat'])
```
